```python
sns.scatterplot(data=df, x="LowNetPrice", y="SoldCount")
```

```python
df["LowNetPrice_under_2500"] = ""
# amplifying the distribution
for key,value in df.iterrows():
    if df.loc[key,"LowNetPrice"] > 2500:
        df.loc[key, "LowNetPrice_under_2500"] = 2500
    else:
        df.loc[key, "LowNetPrice_under_2500"] = df.loc[key,"LowNetPrice"]
```

```python
sns.scatterplot(data=df, y="SoldCount", x="LowNetPrice_under_2500")
```

```python
df.SoldCount.isna().sum()
```

```python
df.SoldCount.fillna(0,inplace=True)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y="SoldCount", x="LowNetPrice_under_2500", data=df)
plt.title("Low Net Price to Sold Count")
plt.show()
```
