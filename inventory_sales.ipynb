```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Inventory_Sales_Dataset.csv")
```

```python
df.head()

```

```python
sns.scatterplot(y="SKU_number", x="ReleaseYear", data= df)
```

```python
df.describe()
```

```python
df[df["PriceReg"] > 2000].sort_values(by="PriceReg", ascending=False)
```

```python
for index, values in df.iterrows():
    if df.loc[index, "PriceReg"] > 500:
        df.loc[index, "PriceReg"] = 500
```

```python
sns.histplot(data=df.PriceReg, bins= 30, kde=False, color='blue')
plt.title('Regular Price Histrogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data=df.MarketingType, kde=False, color='blue')
plt.title('Market Type Histrogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
