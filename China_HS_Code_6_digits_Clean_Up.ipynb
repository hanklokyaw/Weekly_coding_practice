```python
import pandas as pd

df = pd.DataFrame(columns=["CODES", "DESCRIPTION", "YEAR", "UNIT_1_CODES", "UNIT_1_DESCRIPTION", "UNIT_2_CODES", "UNIT_2_DESCRIPTION"])

entries = []

with open("China_hts_6_digits_utf8.csv", "r", encoding="utf8") as f:
    line = f.readline().strip()
    while line:
        if line[:6].isdigit():
            current_entry = line
            while not current_entry.endswith(','):
                current_entry += ' ' + f.readline().strip()
            entries.append(current_entry)
        line = f.readline().strip()

# Lists to store each column's data
CODES = []
DESCRIPTION = []
YEAR = []
UNIT_1_CODES = []
UNIT_1_DESCRIPTION = []
UNIT_2_CODES = []
UNIT_2_DESCRIPTION = []

for entry in entries:
    data = entry.replace('"', '').split(",")
    
    CODES.append(data[0])
    YEAR.append(data[-6])
    UNIT_1_CODES.append(data[-5])
    UNIT_1_DESCRIPTION.append(data[-4])
    UNIT_2_CODES.append(data[-3])
    UNIT_2_DESCRIPTION.append(data[-2])
    DESCRIPTION.append(', '.join(data[1:-6]))  # Anything between 0 and -6, joined by a comma

# Construct the DataFrame using the populated lists.
df['CODES'] = CODES
df['DESCRIPTION'] = DESCRIPTION
df['YEAR'] = YEAR
df['UNIT_1_CODES'] = UNIT_1_CODES
df['UNIT_1_DESCRIPTION'] = UNIT_1_DESCRIPTION
df['UNIT_2_CODES'] = UNIT_2_CODES
df['UNIT_2_DESCRIPTION'] = UNIT_2_DESCRIPTION

print(df)
```
